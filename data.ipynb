```python
import pandas as pd
import glob
```

```python
baseline_csvs = glob.glob('../data/ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_comm/xml/oa_comm_xml.PMC00*xxxxxx.baseline.2023-12-18.filelist.csv')
baseline_dfs = [
    pd.read_csv(baseline_csv) for baseline_csv in baseline_csvs
]
baseline_df = pd.concat(baseline_dfs)
del baseline_dfs
baseline_df
```

```python
daily_update_csvs = glob.glob('../data/ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_comm/xml/oa_comm_xml.incr.*.filelist.csv')
daily_update_dfs = [
    pd.read_csv(daily_update_csv) for daily_update_csv in daily_update_csvs
]
daily_update_df = pd.concat(daily_update_dfs)
del daily_update_dfs
daily_update_df
```

```python
df = pd.concat([baseline_df, daily_update_df])
del baseline_df
del daily_update_df
df = df.drop_duplicates(keep="last")
```

```python
df = df[df["Retracted"] == "no"]
```

```python
df
```

```python
xtract_csvs = glob.glob('output/csv/*.csv')
xtract_dfs = [
    pd.read_csv(xtract_csv) for xtract_csv in xtract_csvs
]
from pathlib import Path
for xtract_df, xtract_csv in zip(xtract_dfs, xtract_csvs):
    xtract_df["query"] = Path(xtract_csv).stem

xtract_df = pd.concat(xtract_dfs)
xtract_df
```

```python
xtract_df_copy = xtract_df[xtract_df['query'] == 'Kallmann syndrome'].copy()
# xtract_df_copy = xtract_df.copy()
xtract_df_copy = xtract_df_copy.drop(columns=['pubmed_article_set'])
xtract_df_copy.columns = ['PMID', 'query']
xtract_df_copy
```

```python
df_query = df.merge(xtract_df_copy, on="PMID", how="inner")
df_query
```

```python
df_query_found = df_query[(~df_query['query'].isna()) & (~df_query['Article File'].isna())]
len(df_query_found['PMID'].unique())
```

```python
df_query_found = df_query[(~df_query['query'].isna())]
len(df_query_found['PMID'].unique())
```

```python
output_articles_dir = Path('../data/articles')
output_articles_dir.mkdir(exist_ok=True, parents=True)

import shutil

from tqdm.contrib.concurrent import process_map

def copy_article(article_found):
    article_found_path = '../data/ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_comm/xml/' + article_found
    try:
        shutil.copy(article_found_path, output_articles_dir)
    except FileNotFoundError:
        # TODO: raise
        pass

process_map(copy_article, df_query_found["Article File"], max_workers=16)
```
